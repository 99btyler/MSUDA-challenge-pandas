# PyCity Schools Analysis

- As the Chief Data Scientist for PyCity, I'm helping the school board and mayor make strategic decisions regarding future budgets and priorities. After analyzing the math and reading scores of every student (plus information on the schools they attended) I discovered trends from which multiple conclusions can be drawn.
- There are 15 schools total with a combined budget of $24.65 million. There are 39,710 total students, of which 65% are passing. 75% are passing math (the average math score being 79) and 86% are passing reading (the average reading score being 82). But where does each school stand? The highest-performing school is Cabrera High School with an overall passing rate of 91% while the lowest-performing school is Rodriguez High School with an overall passing rate of 53%. The schools that spend the most amount of money per student (645-680 dollars) have an overall passing rate of 54%. The schools that spend the least amount of money per student (less than 585 dollars) have an overall passing rate of 90%. The schools with the most students (between 2000 and 5000) have an overall passing rate of 58% while the schools with the least students (less than 1000) have an overall passing rate of 90%. Charter schools have an overall passing rate of 90% while District schools have an overall passing rate of 54%.
- From this, one thing we can see is that spending more per student doesn't guarantee a higher overall passing rate. The percentage of students passing both math and reading is actually lower where spending per student is higher, with the most noticeable difference being in math (a 66% passing rate among students with the most money being spent on them vs. a 93% passing rate among students with the least money being spent on them). Additionally, schools with fewer students have a higher overall passing rate with another noticeable difference being in math (70% passing in schools with 2000-5000 students vs. 94% passing in schools with less than 1000 students). Finally, Charter schools have a higher overall passing rate than District schools with yet another noticable difference in math (94% vs. 67%).
- To get the most growth from spending, money doesn't necessarily have to be increased but rather placed strategically. Since spending more per student doesn't automatically improve scores, making it a priority to maintain a small or medium budget per student will avoid situations where spending is excessive and results are lackluster. Additionally, prioritizing schools with a small or medium student population will reap the most benefit. District schools might be able to keep pace with the overall passing rate of Charter schools if they make math a priority. Finally, larger schools might be able to keep pace with smaller schools if they too make math a priority. Both large schools and District schools might not see improvements just by increasing their budget per student, but prioritizing the lagging subject (math) has the potential to do them wonders!

```python
import pandas as pd
from pathlib import Path

schools_filepath = Path("resources/schools-complete.csv")
students_filepath = Path("resources/students-complete.csv")

schools_data = pd.read_csv(schools_filepath)
students_data = pd.read_csv(students_filepath)

schools_and_students_data = pd.merge(students_data, schools_data, how="left", on="school_name")
schools_and_students_data.head()
```

# Summary of District

```python
num_of_schools = schools_data["School ID"].nunique()
num_of_schools
```

```python
num_of_students = students_data["Student ID"].nunique()
num_of_students
```

```python
budget_sum = schools_data["budget"].sum()
budget_sum
```

```python
math_score_average = students_data["math_score"].mean()
math_score_average
```

```python
reading_score_average = students_data["reading_score"].mean()
reading_score_average
```

```python
num_passing_math = schools_and_students_data[schools_and_students_data["math_score"] >= 70]["Student ID"].nunique()
passing_math_rate = (num_passing_math / num_of_students) * 100
passing_math_rate
```

```python
num_passing_reading = schools_and_students_data[schools_and_students_data["reading_score"] >= 70]["Student ID"].nunique()
passing_reading_rate = (num_passing_reading / num_of_students) * 100
passing_reading_rate
```

```python
num_passing_math_and_reading = schools_and_students_data[(schools_and_students_data["math_score"] >= 70) & (schools_and_students_data["reading_score"] >= 70)]["Student ID"].nunique()
passing_math_and_reading_rate = (num_passing_math_and_reading / num_of_students) * 100
passing_math_and_reading_rate
```

```python
summary_of_district = pd.DataFrame({
    "Total Schools": [num_of_schools],
    "Total Students": [num_of_students],
    "Total Budget": [budget_sum],
    "Average Math Score": [math_score_average],
    "Average Reading Score": [reading_score_average],
    "% Passing Math": [passing_math_rate],
    "% Passing Reading": [passing_reading_rate],
    "% Overall Passing": [passing_math_and_reading_rate]
})

# formatting
summary_of_district["Total Students"] = summary_of_district["Total Students"].map("{:,}".format)
summary_of_district["Total Budget"] = summary_of_district["Total Budget"].map("${:,.2f}".format)

summary_of_district
```

# Summary of School

```python
types_per_school = schools_data.set_index("school_name")["type"]
types_per_school
```

```python
students_per_school = schools_data.set_index("school_name")["size"]
students_per_school
```

```python
budgets_per_school = schools_data.set_index("school_name")["budget"]
budgets_per_school_percapita = budgets_per_school / students_per_school

budgets_per_school_percapita
```

```python
math_scores_per_school = schools_and_students_data.set_index("school_name")["math_score"]
math_scores_per_school_average = math_scores_per_school.groupby("school_name").mean()

math_scores_per_school_average
```

```python
reading_scores_per_school = schools_and_students_data.set_index("school_name")["reading_score"]
reading_scores_per_school_average = reading_scores_per_school.groupby("school_name").mean()

reading_scores_per_school_average
```

```python
num_passing_math = (math_scores_per_school >= 70).count()
num_passing_math_per_school = math_scores_per_school[math_scores_per_school >= 70].groupby("school_name").count()
math_passing_rate_per_school = (num_passing_math_per_school / students_per_school) * 100

math_passing_rate_per_school
```

```python
num_passing_reading = (reading_scores_per_school >= 70).count()
num_passing_reading_per_school = reading_scores_per_school[reading_scores_per_school >= 70].groupby("school_name").count()
reading_passing_rate_per_school = (num_passing_reading_per_school / students_per_school) * 100

reading_passing_rate_per_school
```

```python
num_passing_math_and_reading = schools_and_students_data[(schools_and_students_data["math_score"] >= 70) & (schools_and_students_data["reading_score"] >= 70)]["Student ID"].count()
num_passing_math_and_reading_per_school = schools_and_students_data[(schools_and_students_data["math_score"] >= 70) & (schools_and_students_data["reading_score"] >= 70)].groupby("school_name").size()
math_and_reading_passing_rate_per_school = (num_passing_math_and_reading_per_school / students_per_school) * 100

math_and_reading_passing_rate_per_school
```

```python
summary_of_school = pd.DataFrame({
    "School Type": types_per_school,
    "Total Students": students_per_school,
    "Total School Budget": budgets_per_school,
    "Per Student Budget": budgets_per_school_percapita,
    "Average Math Score": math_scores_per_school_average,
    "Average Reading Score": reading_scores_per_school_average,
    "% Passing Math": math_passing_rate_per_school,
    "% Passing Reading": reading_passing_rate_per_school,
    "% Overall Passing": math_and_reading_passing_rate_per_school
})

# formatting
summary_of_school["Total School Budget"] = summary_of_school["Total School Budget"].map("${:,.2f}".format)
summary_of_school["Per Student Budget"] = summary_of_school["Per Student Budget"].map("${:.2f}".format)

summary_of_school
```

# Highest-Performing Schools (by % Overall Passing)

```python
highest_performing_schools = summary_of_school.sort_values(by="% Overall Passing", ascending=False)
highest_performing_schools.head(5)
```

```python
lowest_performing_schools = summary_of_school.sort_values(by="% Overall Passing")
lowest_performing_schools.head(5)
```

# Scores by School Spending

```python
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_data = summary_of_school.copy()
```

```python
school_spending_data["Spending Ranges (Per Student)"] = pd.cut(budgets_per_school_percapita, bins=spending_bins, labels=spending_labels)

# formatting
school_spending_data["Spending Ranges (Per Student)"] = school_spending_data["Spending Ranges (Per Student)"].astype(str)

school_spending_data
```
